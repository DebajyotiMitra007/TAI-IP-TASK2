```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import io
from google.colab import drive
```

```python
#mounting drive and checking dataset
drive.mount("/content/drive/")
iris = pd.read_csv('/content/drive/MyDrive/project data/Iris.csv')
print("Shape of dataset:", iris.shape)
print("Columns:", iris.columns)
iris.describe()
```

```python
x=iris.iloc[:,:]
x.head()
```

```python
y=iris.iloc[:,-1]
y.head()
```

```python
#to normalize the data
x= preprocessing.StandardScaler().fit_transform(x)
x[0:4]
```

```python
#partition the data
#taking 0.2 giving the 1.0 accuracy . thats why we took 6:4 ratio for train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.4,random_state=1)
y_test.shape
```

```python
#making model
knnmodel = KNeighborsClassifier(n_neighbors=3)
knnmodel.fit(x_train,y_train)
y_predict1=knnmodel.predict(x_test)
```

```python
#accuracy
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_predict1)
print(acc)
```

```python
#confusion Matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test.values,y_predict1)
cm1=pd.DataFrame(data=cm,index=["setosa","versicolor","virginica"],columns=["setosa","versicolor","virginica"])
print(cm1)
```

```python
#out put visualisation
prediction_output=pd.DataFrame(data=[y_test.values,y_predict1],index=['y_test','y_predict1'])
prediction_output.transpose()
```
